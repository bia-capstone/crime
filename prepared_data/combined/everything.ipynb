```python
import pandas as pd, numpy as np

dem_norm = pd.read_csv('output/county_stats_normalized.csv')
dem_cnt = pd.read_csv('output/county_stats_counts.csv')

edu_norm = pd.read_csv('output/education_county_rates.csv').rename(columns={'geo_county':'geo'})
edu_cnt = pd.read_csv('output/education_county_counts.csv').rename(columns={'geo_county':'geo'})

cr_base_rate = pd.read_csv('../crime_records/output/crime_agg_base_rate.csv')
cr_base_count = pd.read_csv('../crime_records/output/crime_agg_base_count.csv')
cr_cat = pd.read_csv('../crime_records/output/crime_agg_category.csv')
cr_name = pd.read_csv('../crime_records/output/crime_agg_name.csv')
cr_cat_rate = pd.read_csv('../crime_records/output/crime_agg_category_rate.csv')
cr_name_rate = pd.read_csv('../crime_records/output/crime_agg_name_rate.csv')

display(dem_norm.head(3), dem_cnt.head(3), edu_norm.head(3), edu_cnt.head(3), cr_base_count.head(3), cr_base_rate.head(3), cr_cat.head(3), cr_name.head(3), cr_cat_rate.head(3), cr_name_rate.head(3))
```

```python
INDEX = ['year', 'county', 'geo']
for col in [i for i in dem_cnt.columns if i not in INDEX]:
    dem_cnt = dem_cnt.rename(columns={col: f'CENSUS_COUNT_{col}'})

for col in [i for i in dem_norm.columns if i not in INDEX]:
    dem_norm = dem_norm.rename(columns={col: f'CENSUS_NORM_{col}'})

dem = dem_cnt.merge(dem_norm, on=INDEX)


for col in [i for i in cr_name.columns if i not in INDEX]:
    cr_name = cr_name.rename(columns={col: f'CRIME_NAME_COUNT_{col}'})

for col in [i for i in cr_cat.columns if i not in INDEX]:
    cr_cat = cr_cat.rename(columns={col: f'CRIME_CATEGORY_COUNT_{col}'})

cr_count = cr_name.merge(cr_cat, on=INDEX)


for col in [i for i in cr_name_rate.columns if i not in INDEX]:
    cr_name_rate = cr_name_rate.rename(columns={col: f'CRIME_NAME_NORM_{col}'})

for col in [i for i in cr_cat_rate.columns if i not in INDEX]:
    cr_cat_rate = cr_cat_rate.rename(columns={col: f'CRIME_CATEGORY_NORM_{col}'})

cr_rate = cr_name_rate.merge(cr_cat_rate, on=INDEX)


for col in [i for i in cr_base_rate.columns if i not in INDEX]:
    cr_base_rate = cr_base_rate.rename(columns={col: f'CRIME_BASE_NORM_{col}'})

for col in [i for i in cr_base_count.columns if i not in INDEX]:
    cr_base_count = cr_base_count.rename(columns={col: f'CRIME_BASE_COUNT_{col}'})

cr_base = cr_base_count.merge(cr_base_rate, on=INDEX)


cr = cr_count.merge(cr_rate, on=INDEX)
cr = cr_base.merge(cr, on=INDEX)

INDEX = ['county', 'geo']
for col in [i for i in edu_cnt.columns if i not in INDEX]:
    edu_cnt = edu_cnt.rename(columns={col: f'EDU_COUNT_{col}'})

for col in [i for i in edu_norm.columns if i not in INDEX]:
    edu_norm = edu_norm.rename(columns={col: f'EDU_NORM_{col}'})

edu = edu_cnt.merge(edu_norm, on=INDEX)

display(
    dem.shape, dem.head(3),
    edu.shape, edu.head(3),
    cr.shape, cr.head(3))
```

```python
df = cr.merge(dem, on=['year', 'county', 'geo'])
df = df.merge(edu, on=['county', 'geo'])
display(df.shape, df.head(3))
df.to_csv('output/everything_grouped.csv', index=False)
```
